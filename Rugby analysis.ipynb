# Rugby points analysis 

In order to understand the difference between Rugby Union and Rugby League, I shall analyse the total score averages across tournaments 8 years back. Undeniably, there shall be a total score disparity between the two types of rugby due to the fact that each variation of rugby has different points assigned to each scoring option. These point diffences shall be discussed in the following section:

**Rugby Union**

**1)** Try = 5 points

**2)** Conversion = 2 points (possible after Try)

**3)** Penalty kick = 3 points

**4)** Drop goal = 3 points


**Rugby League**

**1)** Try = 4 points

**2)** Conversion = 2 points (possible after Try)

**3)** Penalty kick = 2 points

**4)** Drop goal = 1 points

#### Updating column names

```python
#rugby union
dfu.columns = ['year', 'home team', 'away team', 'home total score', 'home FH score', 'home SH score', 'away total score', 'away FH score', 'away SH score']

#rugby league
dfl.columns = ['year', 'home team', 'away team', 'home total score', 'home FH score', 'home SH score', 'away total score', 'away FH score', 'away SH score']
```

### Visualizations

In this section, I focus on the data visualizations. I provide an overview of the development of total match points across seasons as well as first half and second half distribution of points. Additionally, I focus on the points development of the teams that won each season.

####  Visual analysis (Rugby Union)

```python
plt.rcParams["figure.figsize"] = (10,6)

X = ['2014/2015','2015/2016','2016/2017','2017/2018','2018/2019','2019/2020','2020/2021','2021/2022']
Hpoints = dfu_a.groupby(['year'])['home total score'].mean()
Apoints = dfu_a.groupby(['year'])['away total score'].mean()
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Hpoints, 0.4, label = 'Home')
plt.bar(X_axis + 0.2, Apoints, 0.4, label = 'Away')
  
plt.xticks(X_axis, X, rotation=90)
plt.xlabel("Season")
plt.ylabel("Total points")
plt.title("Average total points per season (Rugby Union)")
plt.legend()
plt.show()
```
